# Exploratory Data Analysis

from the dataset we can see that, there are many data and time value which we need to handle

1.date_of_Journey column is in obj form but i need to extract the date and month from that because it could be very useful


```python
train_data['Journey_day'] = pd.to_datetime(train_data.Date_of_Journey,format="%d/%m/%Y").dt.day
```

```python
train_data['Journey_month'] = pd.to_datetime(train_data.Date_of_Journey,format="%d/%m/%Y").dt.month
```

```python
train_data.head()
```

Here we have not taken year because in dataset we are having only one year data

```python
#now the date_of_journey column is not requierd
train_data.drop(['Date_of_Journey'],axis=1,inplace = True)
```

```python
# now we are also having some column which contain the hour and minute data of plane that can also be halpful
#so new extracting the houre and min data from the dep_time
# dep_time is a departure time, when plane leave the gate.

train_data['Dep_hour']=pd.to_datetime(train_data['Dep_Time']).dt.hour

train_data['dep_min']=pd.to_datetime(train_data['Dep_Time']).dt.minute
```

```python
train_data.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
#same work for arrival_time
train_data['Arrival_hour']=pd.to_datetime(train_data['Arrival_Time']).dt.hour

train_data['Arrival_min']=pd.to_datetime(train_data['Arrival_Time']).dt.minute
```

```python
train_data.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
#now the other timedate feature that remain ,from whare we can extract the information is Duration

duration=list(train_data['Duration'])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:
        if "h" in duration[i]:
            duration[i] = duration[i].strip()+" 0m"
        else:
            duration[i] = "0h " + duration[i] 
duration_hours = []
duration_min = []

for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))
    duration_min.append(int(duration[i].split(sep = "m")[0].split()[-1]))
```

```python
#adding this hours and min data to the dataframe
train_data['Duration_hours'] = duration_hours
train_data['duration_min'] = duration_min
```

```python
train_data.drop(['Duration'],axis=1,inplace=True)
```

```python
train_data.head()
```

# Handling Categorical Data

```python
There are many ways to handle the Categorical Variable

some basic ways are...
1. Nominal data = one hot encoding
2. ordinal data = labelEncoding
```

```python
train_data.columns
```

```python
train_data['Airline'].value_counts()
```

```python
# as Airline is a nominal data we will perform the one hot encoding
airline=train_data[['Airline']]
airline=pd.get_dummies(airline,drop_first=True)
airline.head()
```

```python
#now taking second feature = sourse
train_data['Source'].value_counts()
```

```python
# now source data is also a nominal data , we will perform onehot encoding
source=train_data[['Source']]
source=pd.get_dummies(source,drop_first=True)
source.head()
```

```python
# now destination column
train_data['Destination'].value_counts()
```

```python
# now destination data is also a nominal data , we will perform onehot encoding
destination=train_data[['Destination']]
destination=pd.get_dummies(destination,drop_first=True)
destination.head()
```

```python
# now taking the route into the consideration
train_data['Route'].head()
```

```python
# now take total stops into consideration
train_data['Total_Stops'].head()
```

```python
#here we can see that column route and total stops cntain the same information so we can replace it one.
train_data.drop(['Route'],axis=1,inplace=True)
```

```python
train_data['Additional_Info'].value_counts()
```

above we can see that the almost 80% of the values are contained by No info so we can drop that column

```python
train_data.drop(['Additional_Info'],axis=1,inplace=True)
```

```python
train_data['Total_Stops'].value_counts()
```

```python
# it looks likes the case of ordinal variable so i have used a labelencoding
# we can use a sklearns inbuild dunction or can make our own

train_data.replace({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4},inplace =True)
```

```python
train_data.head()
```

```python
# now its time to concatenate the data that we have created and the droping the columns
comb = pd.concat([train_data,airline,source,destination],axis=1)
```

```python
comb.head()
```

```python
comb.drop(['Airline','Source','Destination'],axis=1,inplace=True)
```

```python
# Dep_Time
test_data["Dep_hour"] = pd.to_datetime(test_data["Dep_Time"]).dt.hour
test_data["Dep_min"] = pd.to_datetime(test_data["Dep_Time"]).dt.minute
test_data.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
# Duration
duration = list(test_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))  
```

```python
# Adding Duration column to test set
test_data["Duration_hours"] = duration_hours
test_data["Duration_mins"] = duration_mins
test_data.drop(["Duration"], axis = 1, inplace = True)
```

```python
print("Destination")
print(test_data["Destination"].value_counts())
Destination = pd.get_dummies(test_data["Destination"], drop_first = True)
```
